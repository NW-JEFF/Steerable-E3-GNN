```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv("model_performance/train_L1_loss_only.csv")

y_col_names = ["Baseline (33525) - alpha train MAE", "L1=1e-5 (27371) - alpha train MAE", "L1=1e-3 (33351) - alpha train MAE",
               "L1=5e-3 (18950) - alpha train MAE", "L1=5e-2 (92158) - alpha train MAE", "L1=1e-1 (61219) - alpha train MAE"]
y_labels = ["Baseline", "L1=1e-5", "L1=1e-3",
            "L1=5e-3", "L1=5e-2", "L1=1e-1"]

sns.set(style="whitegrid")
sns.set_context("notebook")

plt.figure(figsize=(6.3, 4.5))

for i, y_col in enumerate(y_col_names):
    sns.lineplot(data=df, x='Step', y=y_col, label=y_labels[i], linewidth=1.5)

plt.xlabel('SGD Step')
plt.xlim(359, 440)
plt.ylabel('Training MAE')
plt.yscale('log')
plt.ylim(0.065, 0.17)

plt.title('Effect of L1 Regularization on Training MAE')
plt.legend(loc='upper right')

ticks = [0.07, 0.08, 0.09, 0.1, 0.11, 0.12, 0.13, 0.15]
tick_labels = [str(tick) for tick in ticks]
plt.gca().yaxis.set_major_formatter(plt.FixedFormatter(tick_labels))
plt.yticks(ticks)

sns.despine()
plt.grid(True, linestyle='--', alpha=0.7)

plt.savefig("/mnt/workspace/linchen/nanxiang/my_segnn/model_performance/train_L1_loss_only.pdf", bbox_inches='tight')

plt.show()
```

```python
df_sparsity = pd.read_csv("model_performance/sparsity_sparsity_test.csv")
df_mae = pd.read_csv("model_performance/mae_sparsity_test.csv")

x_col_names = ["Baseline (33525) - Sparsity", 
               "L1: 1e-3 (0, 0.1, 50) (33351) - Sparsity",
               "L1: 5e-3 (0, 0.1, 50) (18950) - Sparsity",
               "L1: 1e-2 (0, 0.1, 50) (92122) - Sparsity",
               "L1: 5e-2 (0, 0.1, 50) (92158) - Sparsity",
               "L1: 1e-1 (0, 0.1, 50) (61219) - Sparsity"]
y_col_names = ["Baseline (33525) - alpha test MAE", 
               "L1: 1e-3 (0, 0.1, 50) (33351) - alpha test MAE",
               "L1: 5e-3 (0, 0.1, 50) (18950) - alpha test MAE",
               "L1: 1e-2 (0, 0.1, 50) (92122) - alpha test MAE",
               "L1: 5e-2 (0, 0.1, 50) (92158) - alpha test MAE", 
               "L1: 1e-1 (0, 0.1, 50) (61219) - alpha test MAE"]
y_labels = ["Baseline", "L1=1e-3", "L1=5e-3", "L1=1e-2", "L1=5e-2", "L1=1e-1"]

sns.set(style="whitegrid")
sns.set_context("notebook")

custom_colors = ["black", "teal", "blue", "green", "orange", "red"]

plt.figure(figsize=(6.3, 4.3))

for i, y_col in enumerate(y_col_names):
    sns.lineplot(x=df_sparsity[x_col_names[i]], y=df_mae[y_col], label=y_labels[i], linewidth=1.5, color=custom_colors[i])

plt.xlabel('Sparsity')
plt.ylabel('Test MAE')

plt.title('Effect of L1 Pruning on Test MAE')
plt.legend(loc='upper left')

sns.despine()
plt.grid(True, linestyle='--', alpha=0.7)

plt.savefig("/mnt/workspace/linchen/nanxiang/my_segnn/model_performance/mae_vs_sparsity.pdf", bbox_inches='tight')

plt.show()
```
